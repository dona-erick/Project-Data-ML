```python
### les operations de mutliplication, d'addition, de soustraction et de division
tenseur = torch.tensor([1, 2, 3])
addition = tenseur + 10
soust = tenseur -10
multipkication = tenseur * 10
division = tenseur / 10
addition, soust, multipkication, division

```

```python
### ou en utilisant les fonctions add, mul(addition , multi et matmul pour les matrices)
addition = torch.add(tenseur, 10)
soust = torch.sub(tenseur, 10)
multipkication = torch.mul(tenseur, 10)
division = torch.div(tenseur, 10)
multi = torch.matmul(tenseur, tenseur)
addition, soust, multipkication, division, multi
```

```python
### Fonction d'agrégation (max, min sum, mean)

value = torch.arange(start = 1, end = 50, step = 15)
## maximum, minimum
maxi = value.max()
mini = value.min()
moy= value.type(torch.float32).mean()
maxi, mini, moy
```

```python
### ou plus simplement
maxi = torch.max(value)
mini = torch.min(value)
moy = torch.mean(value.type(torch.float32))
maxi, mini, moy
```

```python
### trouver la position d'une valeur
print(value)
position_max = torch.argmax(value)
position_min = torch.argmin(value)
position_max, position_min
```

**Exercice sur les bases**

```python
## Create a random tensor with shape (7, 7)
x = torch.rand(size = (7, 7), dtype =torch.float32)
x.shape, x.ndim, x.device, x.dtype
```

```python
##Perform a matrix multiplication on the tensor from 2 with another random tensor with shape (1, 7)
##(hint: you may have to transpose the second tensor).
y = torch.rand(size = (1, 7), dtype = torch.float32)
y.shape, y.ndim, y.device, y.dtype

mutil = torch.matmul(x, y.T)
mutil
```

```python
mutil = torch.mm(x, y.T)
mutil
```

```python
torch.manual_seed(0)
x = torch.rand(size = (7, 7), dtype =torch.float64)
y = torch.rand(size = (1, 7), dtype = torch.float64)
mutil = torch.mm(x, y.T)
mutil
```

```python
torch.manual_seed(1234)
w = torch.rand(size = (2, 3), dtype =torch.float32)
z = torch.rand(size = (2, 3), dtype = torch.float32)
mutil = torch.mm(w, z.T)
mutil
```

```python
torch.max(mutil), torch.min(mutil)
```

```python
torch.argmax(mutil), torch.argmin(mutil)
```
