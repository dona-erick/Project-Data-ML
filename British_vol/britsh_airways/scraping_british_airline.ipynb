```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
```

```python
import requests

url = "https://www.airlinequality.com/airline-reviews/british-airways"

pages = 100
pages_size = 1000

Avis = []
for x in range(1, pages +1):
    print(f"Scraping the {x}")

    url_create  = f"{url}/page/{x}/?sortby=post_date%3ADesc&pagesize={pages_size}"
    # Collect HTML data from this page
    response = requests.get(url_create)

    # Parse content
    content = response.content
    parsed_content = BeautifulSoup(content, 'html.parser')
    for text in parsed_content.find_all("div", {"class": "text_content"}):
        Avis.append(text.get_text())
    
    print(f"   ---> {len(Avis)} total reviews")
```

```python
df = pd.DataFrame()
df['Avis_airways'] = Avis
df.head()
## sauvegarder dans un fichier
df.to_csv('data_ariline/british_airline.csv', index = False)
```

```python
url_base = "https://www.airlinequality.com/airline-reviews/british-airways"
pages = 1000 ### nombres de pages à pacourir
pages_size = 1000
avis = [] ## les avis ou commentaires
notes = [] 

for x in range (1, pages +1):
    print(f"scraping the page {x}")

    ## url create
    url_create  = f"{url}/page/{x}/?sortby=post_date%3ADesc&pagesize={pages_size}"
    response = requests.get(url_create)
#### recuperation des avis (commentaires)
    if response.status_code ==200:
        content = response.content ### contenu des pages ici ce sont des textes
        soup = BeautifulSoup(content, 'html.parser')
        for avis_compaign in soup.find_all("div", {"class": "text_content"}):
            avis_text = avis_compaign.get_text(strip = True)
            avis.append(avis_text)

### recuperer les notes (stars)
        for rating in soup.find_all('span', {"class": "rating"}):
            rating_notes = rating['class'][1].split("-")[-1]
            notes.append(rating_notes)

        print(f"   ---> {len(avis)} total avis, {len(notes)} total notes")

    else:
        print(f"Failed to retrieve page {x}, status code: {response.status_code}")
        break

# Affichage d'un aperçu des données collectées
print(f"\nCollected {len(avis)} reviews and {len(notes)} ratings.\n")
print("Example review and rating:")
if avis and notes:
    print(f"Review: {avis[0]}")
    print(f"Rating: {notes[0]} stars")

    
```

```python
# Extraction des notes en étoiles (ratings)
url_base = "https://www.airlinequality.com/airline-reviews/british-airways"
pages = 1000 ### nombres de pages à pacourir
pages_size = 1000
avis = [] ## les avis ou commentaires
notes = [] 

for x in range (1, pages +1):
    print(f"scraping the page {x}")

    ## url create
    url_create  = f"{url}/page/{x}/?sortby=post_date%3ADesc&pagesize={pages_size}"
    response = requests.get(url_create)
#### recuperation des avis (commentaires)
    if response.status_code ==200:
        content = response.content ### contenu des pages ici ce sont des textes
        soup = BeautifulSoup(content, 'html.parser')
        rating_spans = soup.find_all("span", {"class": "star-rating"})
        for rating_span in rating_spans:
                # Vérifier si la classe contient une information de note
            classes = rating_span.get("class", [])
            if len(classes) > 1:  # Vérifier qu'il y a bien une deuxième classe
                rating_value = classes[1].split("s")[-1]  # Extraire la note
                notes.append(int(rating_value))  # Convertir en entier

            # Afficher le nombre d'avis et de notes récupérés pour cette page
            print(f"{len(rating_spans)} notes collectées")
```
